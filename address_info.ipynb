```python
import requests
import json

BLOCKFROST_PROJECT_ID = "mainnetp8RL5xwi2RP76SlxXX5deQgImuJLsmTe"

BASE_URL = "https://cardano-mainnet.blockfrost.io/api/v0"

HEADERS = {
    "project_id": BLOCKFROST_PROJECT_ID
}
```

```python
def get_latest_epoch():
    url = f"{BASE_URL}/epochs/latest"
    r = requests.get(url, headers=HEADERS)
    r.raise_for_status()
    return r.json()

epoch = get_latest_epoch()
print("Latest Epoch Information:")
print(json.dumps(epoch, indent=2))
```

```python
def get_tx_metadata_labels():
    url = f"{BASE_URL}/metadata/txs/labels"
    r = requests.get(url, headers=HEADERS)
    r.raise_for_status()
    return r.json()

labels = get_tx_metadata_labels()
print("Transaction Metadata Labels:")
print(json.dumps(labels[:10], indent=2))  # mostramos solo los primeros
```

```python
def get_network_metrics():
    url = f"{BASE_URL}/metrics"
    r = requests.get(url, headers=HEADERS)
    r.raise_for_status()
    return r.json()

metrics = get_network_metrics()
print("Network Metrics:")
print(json.dumps(metrics, indent=2))
```

```python
print(">> Cardano Network Summary")
print(f"Current epoch: {epoch.get('epoch')}")

if len(metrics) > 0:
    latest_metrics = metrics[0]
    print(f"Total transactions: {latest_metrics.get('tx_count')}")
    print(f"Total blocks: {latest_metrics.get('block_count')}")
else:
    print("Network metrics endpoint returned an empty list (expected for new projects).")
```
